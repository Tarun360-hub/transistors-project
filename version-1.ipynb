```python
import pandas as pd
```

```python
df = pd.read_excel('4_parameter_data_vs_mob_rd_rs.xlsx')
df
```

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Define input features and target variable
X = df[['Vd', 'vs', 'mob', 'rd', 'rs']]
y = df['r Pr1.I']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12)

# Create an XGBoost Regressor object
xgb_reg = XGBRegressor(objective='reg:squarederror', n_estimators=1000, learning_rate=0.01, max_depth=18)

# Fit the model on the training data
xgb_reg.fit(X_train, y_train)

# Make predictions on the test set
y_pred = xgb_reg.predict(X_test)

# Calculate Mean Squared Error and R-squared value for the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared Value: {r2}')
```

```python
# Make predictions on the entire dataset
y_full_pred = xgb_reg.predict(X)

# Calculate error (absolute difference) and add to DataFrame
df['r Pr1.I_predicted'] = y_full_pred  # Add predicted values to DataFrame
df['error'] = df['r Pr1.I_predicted'] - df['r Pr1.I']  # Calculate error

# Calculate the average modulus of error (Mean Absolute Error)
average_modulus_of_error = df['error'].abs().mean()
print(f'Average Modulus of Error: {average_modulus_of_error}')

# Display results
df[['r Pr1.I', 'r Pr1.I_predicted', 'error']]
```
